#### What is in a dataframe?

[pd.DataFrame](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html#pandas.DataFrame) is composed of rows and columns

- Each column of a dataframe is a <font color='blue'>pandas.Series</font> object - a dataframe is thus a collection of series:

- A <font color='blue'>pandas.Series</font> is like a column in a table. It is a one-dimensional array holding data of any type.

- data in each column is of the same type. So we say a column is of type <font color='blue'>int64</font> or type <font color='blue'>float64</font>

### Time series

### What is a time series?
[Wiki](https://en.wikipedia.org/wiki/Time_series): In mathematics, a time series is a series of data points indexed (or listed or graphed) in time order. Most commonly, a time series is a sequence taken at successive equally spaced points in time. Thus it is a sequence of discrete-time data.

##### Example:

Dataset of Nobel prize laureates available through an API of the Nobel prize organisation at https://api.nobelprize.org/v1/laureate.csv .

```python
nobel = pd.read_csv("https://github.com/AaltoSciComp/python-for-scicomp/raw/master/resources/data/laureate.csv")
nobel.head()
```

```python
nobel.shape
```

```python
nobel['born']
```

```python
nobel["born"] = pd.to_datetime(nobel["born"], errors ='coerce')
nobel["born"]
```

```python
nobel["died"] = pd.to_datetime(nobel["died"], errors ='coerce')
nobel["year"] = pd.to_datetime(nobel["year"], format="%Y")
```

```python
nobel["lifespan"] = round((nobel["died"] - nobel["born"]).dt.days / 365, 1)
```

```python
nobel['lifespan']
```

```python
nobel.hist(column='lifespan', bins=25, figsize=(8,10), rwidth=0.9)
```

```python
nobel.boxplot(column="lifespan", by="category")
```
